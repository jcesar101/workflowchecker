```python
### Workflow Checker

# Imports
import itertools
import time
import extractData as ed
import computeFitness as cf
import workflowUtils as wu
import numpy as np

from datetime import datetime
from git import Repo
from termcolor import colored

# Constants
REPOS_PATH = "C:\\Tmp\\repos\\260719_0910\\"
startTime = time.time()
fitnessFileNamePrefix = "F:\\fitness_"
avgFitnessFileNamePrefix = "F:\\avgFitness_"
rulesFileNamePrefix = "F:\\rules_"

# Inputs for the algorithm
workflows = ["GitHub","FeatureBranch","GitFlow","Trunk-basedDevelopment","CactusModel"]
repositories = {"GitHub" : ["flow123d","understrap","cppTango"], "FeatureBranch" : ["fflas-ffpack","ePICURE","dragonchain"],
                "GitFlow" : ["Rocket.Chat","swift-sdk","sdl_atf"],
                "Trunk-basedDevelopment" : ["selenium","appium","nusmods"], "CactusModel" : ["ofp","articulate","conveyor"]}
roleBranches = {"GitHub" : {"main" : ["master"], "integration" : [""], "change" : ["others"], "release" : ["master"], "fix" : ["others"]},
                "FeatureBranch" : {"main" : ["master"], "integration" : [""], "change" : ["all"], "release" : [""], "fix" : ["fix"]},
                "GitFlow" : {"main" : ["master"], "integration" : ["develop"], "change" : ["others"], "release" : ["release"], "fix" : ["hotfix","bugfix","fix"]},
                "Trunk-basedDevelopment" : {"main" : ["master"], "integration" : [""], "change" : ["all"], "release" : ["master","release"], "fix" : [""]},
                "CactusModel" : {"main" : ["master","release", "develop"], "integration" : ["all"], "change" : ["all"], "release" : ["master","release","fix"], "fix" : [""]}}
rules = {"GitHub" : ["1.b.i.A","2.a.v.B","3.a.iii.J","3.a.v.E","3.d.iv.A","3.f.i.A","3.f.iii.A","3.i.iii.C","3.i.v.C"],
         "FeatureBranch" : ["1.b.i.A","3.d.iii.A", "3.i.iii.C"],
         "GitFlow" : ["1.b.i.A","1.b.ii.B","1.c.i.A","1.c.ii.A","2.a.ii.D","2.a.v.B","3.a.iv.E","3.d.ii.A","3.e.i.A",
                      "3.e.ii.A","3.f.i.A","3.f.ii.A","4.b.ii.B","4.b.iv.A","4.b.iv.B","4.b.v.A","4.b.v.B","4.c.i","4.d"],
         "Trunk-basedDevelopment" : ["1.b.i.A","3.a.iii.J","3.b.iii.A"],
         "CactusModel" : ["1.b.i.A","3.a.i.J","3.b.iii.B","3.e.i.A","4.c.iv"]}
ruleWeights = {"GitHub" : [1,2,2,1,2,2,2,1,1],
                "FeatureBranch" : [1,1,1],
                "GitFlow" : [1,1,2,2,2,2,2,2,2,2,2,2,2,1,2,1,2,2,2],
                "Trunk-basedDevelopment" : [1,1,1],
                "CactusModel" : [1,1,2,1,1]}
parameters = {"GitHub" : {"freqThreshold" : 2}, "FeatureBranch" : {},
              "GitFlow" : {}, "Trunk-basedDevelopment" : {"freqThreshold" : 2}, "CactusModel" : {"freqThreshold" : 2}}

# Inputs for plotting and displaying
colorTerm = {"GitHub" : "red", "FeatureBranch" : "blue", "GitFlow" : "green","Trunk-basedDevelopment" : "magenta",
             "CactusModel" : "yellow"}
patternTerm = {"GitHub" : '-', "FeatureBranch" : '|', "GitFlow" : '\\', "Trunk-basedDevelopment" : '.',
               "CactusModel" : 'x'}

# Output from the algorithm
fitness = dict()

# Outputs for plotting and saving
reposAvgFitness = dict()
tmpAvgFitness = dict()
ruleValues = dict()

# Get unique set of rules


# Algorithm
for wfEval in workflows:
    maxFitness = np.sum(np.abs(ruleWeights[wfEval]))
    fitness[wfEval] = dict()
    reposAvgFitness[wfEval] = dict()
    tmpAvgFitness[wfEval] = dict()
    ruleValues[wfEval] = dict()
    print(colored("\nFor " + wfEval + ": ", colorTerm[wfEval]), end = '')
    for wfRepo in workflows:
        avgFitness = 0
        ruleValues[wfEval][wfRepo] = dict()
        for repo in range(0, len(repositories[wfRepo])):
            ruleValues[wfEval][wfRepo][repositories[wfRepo][repo]] = dict()
            wfFitness = 0
            localPath = REPOS_PATH + wfRepo + "\\" + repositories[wfRepo][repo]

            # Repository location
            localRepo = Repo(localPath)
            
            # Evaluate rules on each repository
            for rule, weight in zip(rules[wfEval], ruleWeights[wfEval]):
                ruleFitness = getattr(cf,"rule_" + rule.replace(".","_"))(localRepo,weight,roleBranches[wfRepo],
                                                                         parameters[wfEval])
                wfFitness += ruleFitness
                ruleValues[wfEval][wfRepo][repositories[wfRepo][repo]][rule] = ruleFitness
                #print("Rule " + rule.replace(".","_") + " fitness: " + str(wfFitness))

            # Compute repository fitness and aggregate workflow fitness
            wfFitness /= maxFitness
            avgFitness += wfFitness

            # Store and display fitness
            fitness[wfEval][repositories[wfRepo][repo]] = wfFitness
            if(wfRepo == wfEval):
                print(colored(repositories[wfRepo][repo] + ": " + str("{:.2%}".format(wfFitness)) + " ", colorTerm[wfEval]),
                      end = '')
            else:
                print(repositories[wfRepo][repo] + ": " + str("{:.2%}".format(wfFitness)) + " ", end = '')

        # Compute average fitness per evaluated workflow
        avgFitness /= len(repositories[wfRepo])
        reposAvgFitness[wfEval][wfRepo] = avgFitness
        
        # Display the fitness per repository and the average fitness for group of repositories following the same workflow
        for repo2 in range(0, len(repositories[wfRepo])):
            tmpAvgFitness[wfEval][repositories[wfRepo][repo2]] = avgFitness
        if(wfRepo == wfEval):
            print(colored("avg: " + str("{:.2%}".format(avgFitness)) + " ", colorTerm[wfEval]), end = '')
        else:
            print("avg: " + str("{:.2%}".format(avgFitness)) + " ", end = '')

print("\n\nExecution time: %s seconds" % (time.time() - startTime))

# Save fitness, averages and rule values for all repos in CSV files
timestamp = datetime.now().strftime("%Y%m%d%H%M%S")
wu.saveFitness(fitness, repositories, fitnessFileNamePrefix, timestamp)
wu.saveFitness(reposAvgFitness, repositories, avgFitnessFileNamePrefix, timestamp)
wu.saveRules(ruleValues, repositories, rules, rulesFileNamePrefix, timestamp)

# Plot fitness and averages, and save PNG image
wu.plotFitness(fitness, tmpAvgFitness, workflows, repositories, colorTerm, patternTerm, fitnessFileNamePrefix, timestamp)
```
